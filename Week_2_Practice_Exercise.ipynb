```python
# Exercise 8: Estimate best-fit line (Not taught in class)
x_vals = X.reshape(-1)
x_vals
```

```python
y_vals = y.reshape(-1)
y_vals
```

```python
xmean = x_vals.mean()
ymean = y_vals.mean()
```

```python
print('xmean :' , xmean)
print('ymean :' , ymean)
```

```python
sumProd = 0.0
sqSum = 0.0

length = len(x_vals)

for i in __builtins__.range(length):
    #print(x_vals[i], y_vals[i])

    prod = (x_vals[i] - xmean) * (y_vals[i] - ymean)
    #print(prod)
    sumProd += prod

    sqX = (x_vals[i] - xmean) ** 2
    sqSum += sqX

print('sumProd :' , sumProd)
print('sqSum :' , sqSum)

w = sumProd / sqSum
print("w is : ", w)

b = ymean - w * xmean
print("b is: ", b)

for i in __builtins__.range(len(x_vals)):
    y_pred = w * x_vals[i] + b
    print("Y Preds" , y_pred)
```

```python
# Exercise 10: Shape-mismatch debugging (challenge
u = np.array([1.0, 2.0, 3.0]) # (3,)
V = np.array([[1.0, 2.0],
              [3.0, 4.0]]) # (2,2)
print(u @ V)
```

```python
# the error is that the no of element in 1D array, should be same as # of rows in the 2D array..
# fix is to add one more row to the V array..
u = np.array([1.0, 2.0, 3.0]) # (3,)
V = np.array([[1.0, 2.0],
              [3.0, 4.0],
              [6.0, 8.0]]) # (3,2)
print(u @ V)
```
